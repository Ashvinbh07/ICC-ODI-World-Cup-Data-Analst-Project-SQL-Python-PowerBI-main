```python
import pandas as pd
```

```python
df_final = pd.read_csv('cleaned_data/final_worldcup_dataset.csv')
df_match = pd.read_csv('cleaned_data/master_matches.csv')
```

```python
from sqlalchemy import create_engine
```

```python
files = {
    "final_worldcup_dataset": df_final,
    "master_matches": df_match
}
```

```python
engine = create_engine("mysql+pymysql://root:password@localhost:3306/icc_odi_worldcup")
```

```python
for table_name, df in files.items(): df.to_sql(name=table_name,con=engine,if_exists='replace',index=False)
```
